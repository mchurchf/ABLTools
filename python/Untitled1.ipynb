```python
# viewBoundaryData.py
#
#
# Matt Churchfield
# National Renewable Energy Laboratory
# 7 June 2018
#



from inflowPrepMMC import inflowPrepMMC
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt




rootDir = '/scratch/joniko/onshore/onshore_10km/postProcessing/boundaryData'
boundaries = ['east']
fieldName = 'U'
pointsFileName = 'points'
#time = 20000
#time = 20000.3369272
time = 20060.1135715







#for i in range(len(boundaries)):
for i in range(1):
    boundaryDir = rootDir + '/' + boundaries[i] + '/'
    
    boundary = inflowPrepMMC()
    
    boundary.readBoundaryDataPoints(boundaryDir + pointsFileName)
    
    u_ = boundary.readBoundaryDataField(boundaryDir + '/' + str(time) + '/' + fieldName)
    T_ = boundary.readBoundaryDataField(boundaryDir + '/' + str(time) + '/' + 'T')
    k_ = boundary.readBoundaryDataField(boundaryDir + '/' + str(time) + '/' + 'k')
    u = u_[:,0]
    v = u_[:,1]
    w = u_[:,2]
    uh = np.sqrt(np.square(u) + np.square(v))
    T = T_[:,0]
    k = k_[:,0]
    
    print(T.shape)
    print(k.shape)
    print(u_.shape)
    
    plt.figure(1,figsize=(20,8))
    #plt.scatter(boundary.xyz[:,1], boundary.xyz[:,2], boundary.xyz[:,2], cmap='nipy_spectral')
    plt.scatter(boundary.xyz[:,1], boundary.xyz[:,2], s=0.75, c=uh, cmap='RdBu_r')
    plt.colorbar()
    #plt.axis('equal')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    #plt.show()
    
    plt.figure(2,figsize=(20,8))
    #plt.scatter(boundary.xyz[:,1], boundary.xyz[:,2], boundary.xyz[:,2], cmap='nipy_spectral')
    plt.tricontourf(boundary.xyz[:,1], boundary.xyz[:,2], uh, vmin=14.0, vmax=15.5,  levels=201, cmap='RdBu_r')
    plt.colorbar()
    #plt.axis('equal')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    #plt.show()
    
    plt.figure(3,figsize=(20,8))
    #plt.scatter(boundary.xyz[:,1], boundary.xyz[:,2], boundary.xyz[:,2], cmap='nipy_spectral')
    plt.tricontourf(boundary.xyz[:,1], boundary.xyz[:,2], T, levels=41, cmap='plasma')
    plt.clim([14.5,15.5])
    plt.colorbar()
    #plt.axis('equal')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    #plt.show()
    
    plt.figure(4,figsize=(20,8))
    #plt.scatter(boundary.xyz[:,1], boundary.xyz[:,2], boundary.xyz[:,2], cmap='nipy_spectral')
    plt.tricontourf(boundary.xyz[:,1], boundary.xyz[:,2], k, levels=41, cmap='Greys_r')
    plt.colorbar()
    #plt.axis('equal')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    #plt.show()
    
```
